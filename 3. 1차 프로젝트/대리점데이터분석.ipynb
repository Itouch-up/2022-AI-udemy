```python
import pandas as pd
user_data = pd.read_csv('./data/대리점데이터/user.csv')
customer_data = pd.read_excel('./data/대리점데이터/customer_data.xlsx')
```

```python
user_data['purchase_date'] = pd.to_datetime(user_data['purchase_date'])
user_data['purchase_month'] = user_data['purchase_date'].dt.strftime('%Y%m')
res = pd.pivot_table(data = user_data, index = 'purchase_month', columns = 'item_name', aggfunc = 'count', fill_value = 0)
res
```

## user_data 오류 수정

```python
user_data['item_name'] = user_data['item_name'].str.upper() # 대문자로 수정
user_data['item_name'] = user_data['item_name'].str.replace(" ","") # 띄어쓰기가 있는 것을 없는 것으로 수정
user_data.head()
```

```python
print(user_data['item_name'].unique())
print(len(user_data['item_name'].unique()))
```

```python
# 1. 다른데서 찾아서 넣을 수 있는가?
# 2. 다른 값으로 채워도 괜찮은가?
# 3. 불가능 하다면 삭제
user_data
```

```python
# 최대값과 최소값 평균이 같다는 말은 결국 가격이 변한적이 없고, 각 상품은 같은 가격을 가르킨다.
mean_item_name = user_data.groupby('item_name', as_index = False).agg(max = ('item_price', 'max'),
                                    min = ('item_price', 'min'),
                                    mean = ('item_price', 'mean'))
mean_item_name.head()
```

```python
drop_item_price = user_data.drop('item_price', axis = 1) # user_data에 있는 item_price column을 제거
mean_item_name_df = mean_item_name[['item_name','mean']] # mean_item_name에 있는 칼럼 중 ['item_name','mean'] 만 선택 
merge_data = pd.merge(drop_item_price, mean_item_name_df, on = 'item_name', how = 'left') # drop_item_price, mean_item_name_df merge
user_data = merge_data.rename(columns = {'mean':'item_price'}) # column 명을 mean -> item_price
```

## customer_data 오류 수정

```python
# 고객 이름
customer_data['고객이름'] = customer_data['고객이름'].str.replace(' ','')
```

```python
customer_data['등록일'][0]
```

```python
customer_data['등록일'][1]
```

```python
# isdigit 문자열 중 숫자로 표현 되어 있는 데이터 들을 찾는다.
flg_is_serial = customer_data['등록일'].astype('str').str.isdigit()
```

### loc vs iloc

```python
# loc -> index, column name
# iloc -> data number
customer_data.loc[3,'지역']
customer_data.iloc[3,1]
```

```python
customer_data.loc[4, '고객이름'] # 김시온
customer_data.iloc[1,1] # E시
customer_data.loc[1:3,'고객이름'] # 김도윤, 김지한, 김하윤
customer_data.iloc[1:3,0] # 김도윤, 김지한
```

### loc를 활용해서 데이터 선택

```python
# 컴퓨터에서 시간을 표현할 때 대체로 1900년도 1월 1일을 기준으로 합니다.
# timedelta는 시간의 범위, datetime은 특정 시간을 의미합니다.
# '~'은 조건에 해당 되지 않는 데이터를 가져옵니다.
fromSerial = pd.to_timedelta(customer_data.loc[flg_is_serial, '등록일'].astype('float'), unit = 'D') + pd.to_datetime('1900/01/01')
fromString = pd.to_datetime(customer_data.loc[~flg_is_serial, "등록일"])
customer_data['등록일'] = pd.concat([fromSerial,fromString])
customer_data
```
